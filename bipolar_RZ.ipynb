```python
import numpy as np
import matplotlib.pyplot as plt

def bipolar_return_to_zero(bits, voltage=1, bit_duration=1):
    signal = []
    last_voltage = 0  # Initialize the last voltage to 0
    for bit in bits:
        if bit == 1:
            voltage_level = voltage if last_voltage == 0 else -voltage
            signal.extend([voltage_level, voltage_level])  # High to Low transition for 1
            last_voltage = voltage_level
        else:
            signal.extend([0, 0])  # No transition for 0

    time = np.arange(0, len(bits) * bit_duration * 2, bit_duration)
    plt.step(time, signal)
    plt.title('Bipolar Return-to-Zero Line Coding')
    plt.xlabel('Time')
    plt.ylabel('Voltage')
    plt.ylim(-1.5, 1.5)  # Adjust y-axis limit for better visualization
    plt.grid(True)  # Enable grid
    plt.show()

# Example usage
bipolar_return_to_zero([1, 0, 1, 1, 0])
```
