```python
import numpy as np
import matplotlib.pyplot as plt

def unipolar_nrz(bits, voltage=1, bit_duration=1):
    signal = []
    for bit in bits:
        signal.extend([voltage] if bit == 1 else [0] * int(bit_duration))

    time = np.arange(0, len(bits) * bit_duration, 1)
    plt.step(time, signal)
    plt.title('Unipolar NRZ Line Coding')
    plt.xlabel('Time')
    plt.ylabel('Voltage')
    plt.show()

# Example usage
unipolar_nrz([1, 0, 1, 1,0])
```
