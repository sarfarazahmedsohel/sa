```python
import numpy as np
import matplotlib.pyplot as plt

def polar_nrz(bits, voltage=1, bit_duration=1):
    signal = [voltage if bit == 1 else -voltage for bit in bits]

    time = np.arange(0, len(bits) * bit_duration, 1)
    plt.step(time, signal)
    plt.title('Polar NRZ Line Coding')
    plt.xlabel('Time')
    plt.ylabel('Voltage')
    plt.show()

# Example usage
polar_nrz([1, 0, 1, 1,0])
```
