```python
import numpy as np
import matplotlib.pyplot as plt

def unipolar_return_to_zero(bits, voltage=1, bit_duration=1):
    signal = []
    for bit in bits:
        if bit == 1:
            signal.extend([voltage, voltage])  # High to Low transition for 1
        else:
            signal.extend([0, 0])  # No transition for 0

    time = np.arange(0, len(bits) * bit_duration * 2, bit_duration)
    plt.step(time, signal)
    plt.title('Unipolar Return-to-Zero Line Coding')
    plt.xlabel('Time')
    plt.ylabel('Voltage')
    plt.ylim(-0.5, 1.5)  # Adjust y-axis limit for better visualization
    plt.grid(True)  # Enable grid
    plt.show()

# Example usage
unipolar_return_to_zero([1, 0, 1, 1, 0])
```
