```python
import numpy as np
import matplotlib.pyplot as plt

def manchester(bits, voltage=1, bit_duration=1):
    signal = []
    for bit in bits:
        signal.extend([voltage, -voltage] if bit == 1 else [-voltage, voltage])

    time = np.arange(0, len(bits) * bit_duration, 0.5)
    plt.step(time, signal, where='post')
    plt.title('Manchester Line Coding')
    plt.xlabel('Time')
    plt.ylabel('Voltage')
    plt.show()

# Example usage
manchester([1, 0, 1, 1,0])
```
